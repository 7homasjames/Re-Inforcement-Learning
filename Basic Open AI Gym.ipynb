### Import Packages

```python
import numpy as np
import gym

```

### Create Enviornment

```python
env=gym.make("FrozenLake-v1")
obs=env.observation_space.n
act=env.action_space.n
obs
```

### Reset the enviornment

```python
state=env.reset()
```

### Run The enviornment For n Steps

```python
episodes=10
timestep=10
```

```python
for i in range(episodes):
    env.reset()
    for t in range(timestep):
        env.render()#Visualize the current enviornment
        action=env.action_space.sample()#Select a random action
        new_state,reward,done,info=env.step(action)#Takes action

        if done==True:
            break
```

### Close the Enviornment

```python
env.close()
```
