## Multi Armed Bandit Problem using Thompson Sampling

### Scenario

Imagine that you are at a casino, and that you have $1,000 to play the slot machines. There are six slot machines available, and each turn playing a machine costs $1. The probability of winning on any given turn (which is called the conversion rate) is unknown, and varies from machine to machine. In order to maximize your chances of winning, you would like to be able to identify the slot machine that has the highest conversion rate as quickly as possible. How can you achieve this goal?

### Import Libraries

```python
import numpy as np
```

### Define Enviornment

```python
no_of_turns=1000 #We have 1000$ and each turn cost 1$
no_of_slot_machines=6

# To keep track of wins and loses
no_of_positive_rewards=np.zeros(no_of_slot_machines)
no_of_negative_rewards=np.zeros(no_of_slot_machines)

np.random.seed(33) #To ensure results are reproducable

no_of_positive_rewards
```

#### Creating a random conversion rate between 1% & 15%

```python
conv_rate=np.random.uniform(0.01,0.15,no_of_slot_machines)
conv_rate
```

```python
for i in range(6):
    print("Conversion rate of slot machine {0}: {1}%".format(i,conv_rate[i]))
```

```python
for i in range(6):
    print("Conversion rate of slot machine {0}: {1:%}".format(i,conv_rate[i]))
```

### Create Dataset & Random Sampling

The data set is a matrix with one row for each turn, and one column for each slot machine.Each item in the matrix represents the outcome of what would happen if we were to play a particular slot machine on that particular turn. A value of "1" indicates that we would win, while a value of "0" indicates that we would lose. The number of "wins" for each slot machine is determined by its conversion rate.

```python
np.random.rand()
```

```python
np.random.seed(55)

outcome=np.zeros((no_of_turns,no_of_slot_machines))
#Get a random number between 0.0 and 1.0.
#If the random number is less than or equal to this slot machine's conversion rate, then set the outcome to "1".
#Otherwise, the outcome will be "0" because the entire matrix was initially filled with zeros.

for i in range(no_of_turns):
    for j in range(no_of_slot_machines):
        if np.random.rand()<= conv_rate[j]:
            outcome[i][j]=1
            

    for i in range(6):
        print(outcome[0:15,0:6])#display
            

        
```

#### Mean of conversion rate

```python
for i in range(6):
    print("Mean for coloumn {0}: {1:%}".format(i,np.mean(outcome[:,i])))
```

### Thompson Sampling

```python
for i in range(no_of_turns):
    index_of_machine_to_play=-1
    max_beta=-1
    
    for j in range(no_of_slot_machines):
        a=no_of_positive_rewards[j]+1
        b=no_of_negative_rewards[j]+1
        
        random_beta=np.random.beta(a,b)
        
        if random_beta>max_beta:
            max_beta=random_beta
            index_of_machine_to_play=j
    
    if outcome[i][j]==1:
        no_of_positive_rewards[index_of_machine_to_play] +=1
    else:
        no_of_negative_rewards[index_of_machine_to_play] +=1
        

        
```

#### Compute and display the total number of times each slot machine was played

```python
number_of_times_played = no_of_positive_rewards + no_of_negative_rewards 
for i in range(no_of_slot_machines): 
    print('Slot machine {0} was played {1} times'.format(i, number_of_times_played[i]))

#identify and display the best slot machine to play
print('\nOverall Conclusion: The best slot machine to play is machine {}!'.format(np.argmax(number_of_times_played)))
```

### Comparison

```python
#compute total number of wins using Thompson Sampling strategy
total_wins_thompson_sampling = np.sum(no_of_positive_rewards)

#determine how many times we would win if we randomly choose a slot machine to play for each turn
total_wins_random_sampling = 0
for i in range(no_of_turns):
  index_of_machine_to_play = np.random.randint(0, no_of_slot_machines) #randomly choose a machine to play
  if outcome[i][index_of_machine_to_play] == 1:
    total_wins_random_sampling += 1

#display results
print('Total wins with Thompson Sampling: {0:.0f}'.format(total_wins_thompson_sampling))
print('Total wins with Random Sampling: {0:.0f}'.format(total_wins_random_sampling))
```
